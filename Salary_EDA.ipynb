### How many Job Title are there? and for which Job Title most and less salary is given

```python
salary['Job Title'].nunique()
```

There are 191 job titles

```python
# Find top 10 according to their number of records
salary['Job Title'].value_counts()[:10]
```

```python
salary['Job Title'].value_counts()[:10].index
```

```python
sns.barplot(x= salary['Job Title'].value_counts()[:10].index,
            y= salary['Job Title'].value_counts()[:10].values)
plt.xticks(rotation=90)
plt.xlabel('Job Title')
plt.ylabel('Total Number of Records')
plt.title('Top 10 Job Title according to their number of records')
plt.show()
```

```python
# Now, finding top 10 max and min salary Job Title
max_salary_job_title = salary.groupby('Job Title')['Salary'].max().reset_index()
min_salary_job_title = salary.groupby('Job Title')['Salary'].min().reset_index()
```

```python
max_salary_job_title = max_salary_job_title.sort_values(by=['Salary'], ascending=False)
max_salary_job_title = max_salary_job_title.head(10)
min_salary_job_title = min_salary_job_title.sort_values(by=['Salary'])
min_salary_job_title = min_salary_job_title.head(10)
```

```python
sns.barplot(x='Job Title' , y='Salary' , data=max_salary_job_title)
plt.xticks(rotation=90)
plt.title('Top 10 Job Titles those offer max salary')
plt.show()
```

it means, maximum salary is provided to CTO of companies

```python
sns.barplot(x='Job Title' , y='Salary' , data=min_salary_job_title)
plt.xticks(rotation=90)
plt.title('Top 10 Job Titles those offer min salary')
plt.show()
```

it means, min salary is offered to junior business Operations Analyst

### Is higher Education offers more Salary?

```python
salary['Education Level'].unique()
```

First we need to clean this data
- Bachelor's and Bachelor's Degree are same.
- Master's and Master's Degree are same.
- PhD and phD are same.

```python
# Data Cleaning for Education Level column
salary['Education Level'] = salary['Education Level'].replace({"Bachelor's Degree": "Bachelor's",
                                                           "Master's Degree": "Master's",
                                                           "PhD" : "phD"})
salary
```

```python
# First represent their count of records of Education level
sns.barplot(x=salary['Education Level'].value_counts().index,
            y=salary['Education Level'].value_counts().values)
plt.title('Education Level Records present')
plt.xlabel('Education Level')
plt.ylabel('Number of Record Present')
plt.show()
```

it means, most records are present for Bachelor's

```python
# Represent their count of Education level based on their genders
sns.countplot(x='Education Level' , data=salary, hue='Gender')
plt.title('Education Level Records present')
plt.xlabel('Education Level')
plt.ylabel('Number of Record Present')
plt.show()
```

- More Females are there in Masters

```python
# Find relation between Education level and Salary
salary_education_level = salary.groupby('Education Level')['Salary'].max().reset_index()
sns.barplot(x='Education Level',y='Salary',data=salary_education_level)
plt.show()
```

it means, most salary is given to Bachelor's degree peoples
